```python
import matplotlib.pyplot as plt
import numpy as np

# Defining labels and values for Time 0 and Time + 5
labels=np.array(['Rivalry', 'Substitute Threat', 'Supplier Power', 'Buyer Power', 'Entry Threat'])
stats=np.array([4, 2, 2, 2, 3])  # Time 0 values (1=Low, 5=High)
stats2=np.array([4, 3, 2.5, 3, 3.5])  # Time + 5 values (1=Low, 5=High)

# Create radar chart
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
stats=np.concatenate((stats,[stats[0]]))
stats2=np.concatenate((stats2,[stats2[0]]))
angles+=angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='green', alpha=0.25)
ax.fill(angles, stats2, color='red', alpha=0.25)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Plot Time 0 data
ax.plot(angles, stats, color='green', linewidth=2, linestyle='solid', label='Time 0')
ax.fill(angles, stats, color='green', alpha=0.25)

# Plot Time + 5 data
ax.plot(angles, stats2, color='red', linewidth=2, linestyle='solid', label='Time + 5')
ax.fill(angles, stats2, color='red', alpha=0.25)

# Add legend and title
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
ax.set_title('H&M Comparative Industry Structure Analysis')

plt.show()
```
